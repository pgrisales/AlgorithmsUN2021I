# Fibonacci Algorithms Comparison

```python
import matplotlib.pyplot as plt
import numpy as np 
import time
```

## Recursive

```python
def fib_recursive(n):
    if(n <= 1):
      return n
    else:
      return fib_recursive(n-1) + fib_recursive(n-2)
```

```python
%%time 

n = 20  # USING 50 PRODUCES A STACK OVERFLOW
x_values = []
y_values = []
for i in range(1, n+1):
  x_values.append(i)

  start_time = time.time()
  fib_recursive(i)
  y_values.append(time.time() - start_time)

plt.plot(x_values, y_values)
```

## Iterative Linear

```python
def fib_iterative_linear(n):
  fib_numbers = [0] * (n+1)

  fib_numbers[0] = 0
  fib_numbers[1] = 1

  for i in range(2, n+1):
    fib_numbers[i] = fib_numbers[i-1] + fib_numbers[i-2]
  
  return fib_numbers[n]
```

```python
n = 1000  # USING 50 PRODUCES A STACK OVERFLOW
x_values = []
y_values = []
for i in range(1, n+1):
  x_values.append(i)
  start_time = time.time()
  fib_iterative_linear(i)
  y_values.append(time.time() - start_time)

plt.plot(x_values, y_values)
```

## Iterative Logarithmic

```python
# Taken from https://www.geeksforgeeks.org/program-for-nth-fibonacci-number/
# Method 4 optimized

# Fibonacci Series using
# Optimized Method
def fib_log(n):
	
	F = [[1, 1],
		[1, 0]]
	if (n == 0):
		return 0
	power(F, n - 1)
		
	return F[0][0]
	
def multiply(F, M):
	
	x = (F[0][0] * M[0][0] +
		F[0][1] * M[1][0])
	y = (F[0][0] * M[0][1] +
		F[0][1] * M[1][1])
	z = (F[1][0] * M[0][0] +
		F[1][1] * M[1][0])
	w = (F[1][0] * M[0][1] +
		F[1][1] * M[1][1])
	
	F[0][0] = x
	F[0][1] = y
	F[1][0] = z
	F[1][1] = w
		
# Optimized version of
# power() in method 4
def power(F, n):

	if( n == 0 or n == 1):
		return;
	M = [[1, 1],
		[1, 0]];
		
	power(F, n // 2)
	multiply(F, F)
		
	if (n % 2 != 0):
		multiply(F, M)

# This code is contributed
# by ChitraNayal
```
