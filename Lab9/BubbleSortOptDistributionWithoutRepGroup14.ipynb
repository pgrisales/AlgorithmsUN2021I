# Bubble Sort Without Repetition

```python
import random
from itertools import permutations

# Permutations without repetition
n = 7
iterable_values = [random.randrange(1, 4, 1) for i in range(n)]
unique_iterable_values = set(iterable_values)    # Ensures no repetition
perm = permutations(iterable_values)

perm = list(perm)

ls =  [] 
c = 0
# Print the obtained permutations
for i in list(perm):
    print(i)
    c += 1

print(c)
```

```python
def bubbleSortOptimizedAllCounters(arr):

    comps = 0
    swaps = 0
    steps = 0
    passes = 0 
    n = len(arr)

    # Traverse through all array elements
    while(n > 0):
        newn = 0    # Optimizes it
        passes +=1
        # Last i elements are already in place
        for j in range(n-1):

            comps +=1
            if arr[j] > arr[j+1] :
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swaps +=1
                newn = j + 1
        n = newn
    steps = comps + swaps    
  
    return(comps,swaps,steps,passes)

# Wrapping function
def bubbleSortAllCounters(arr):
  return bubbleSortOptimizedAllCounters(arr)

bubbleSortOptimizedAllCounters(list(perm[0]))
```

```python
from itertools import permutations
import matplotlib.pyplot as plt
import numpy as np 
import math

comps = []
swaps = []
steps = []
passes = [] 

# Print the obtained permutations
for i in list(perm):
    i = list(i)
    c,sw,st,p = bubbleSortAllCounters(i)
    comps.append(c)
    swaps.append(sw)
    steps.append(st)
    passes.append(p)


b = np.array(range(min(comps)-1,max(comps)+2) ) +0.5
plt.hist(comps,bins=b, color="red")
plt.title("Bubble Sort")
plt.xlabel("Comparisons ")
plt.ylabel("Frequency")
plt.show()

b = np.array(range(min(swaps)-1,max(swaps)+2) ) +0.5
plt.hist(swaps,bins=b,color="blue")
plt.title("Bubble Sort")
plt.xlabel("Swaps ")
plt.ylabel("Frequency")
plt.show()

b = np.array(range(min(steps)-1,max(steps)+2) ) +0.5
plt.hist(steps,bins=b,color="green")
plt.title("Bubble Sort")
plt.xlabel("Steps ")
plt.ylabel("Frequency")
plt.show()

b = np.array(range(min(passes)-1,max(passes)+2) ) +0.5
plt.hist(passes,bins=b,color="brown")
plt.title("Bubble Sort")
plt.xlabel("Passes")
plt.ylabel("Frequency")
plt.show()
    
    
print("n",n)  
print("number of permutations n! =",math.factorial(n)) 
print()

print("Comps") 
print("Comps formula (n-1)*n/2 =",(n-1)*n/2)
print("Comps exper min =",min(comps))
print("Comps exper max =",max(comps))
print("Comps exper mean =",np.mean(comps))
print("Comps exper stdev =",np.std(comps))
print()

print("Swaps") 
print("Swaps exper min =",min(swaps))
print("Swaps min formula = 0")
print("Swaps exper max =",max(swaps))
print("Swaps max formula (n-1)*n/2 =",(n-1)*n/2)
print("Swaps exper mean =",np.mean(swaps))
print("Swaps mean formula (n-1)*n/4 =",(n-1)*n/4)
print("Swaps exper stdev =",np.std(swaps))
print()

print("Steps") 
print("Steps exper min =",min(steps))
print("Steps min formula (n-1)*n/2 =",(n-1)*n/2)
print("Steps exper max =",max(steps))
print("Steps max formula (n-1)*n =",(n-1)*n)
print("Steps exper mean =",np.mean(steps))
print("Steps mean formula 3/4*n**2+17/4*n-4 =",(3/4)*(n**2-n))
print("Steps exper stdev",np.std(steps))
print()

print("Passes") 
print("Passes formula (n-1) =",(n-1))
print("Passes exper min =",min(passes))
print("Passes exper max =",max(passes))
print("Passes exper mean =",np.mean(passes))
print("Passes exper stdev",np.std(passes))
```
