## 1. Maximum Amount of Gold

```python
def sol(W, n, w):

    value = [[0]*(len(w)+1) for _ in range(W+1)]

    for i in range(1,len(w)+1):
        for j in range(1,W+1):
            value[j][i] = value[j][i-1]
            if w[i-1] <= j:
                val = value[j-w[i-1]][i-1] + w[i-1]
                if val >= value[j][i]:
                    value[j][i] = val
    print(value[W][len(w)])

if __name__ == '__main__':
    W, n = [int(i) for i in input().split()]
    w = [int(i) for i in input().split()]
    sol(W,n,w)
```

## 2. Partitioning Souvenirs

```python
import numpy as np

def solution(n, iw):
    tw = sum(iw)
    if n<3: 
        print('0')
    elif tw % 3 != 0:
        print('0')
    else:
        W = tw//3

        count = 0 
        value = np.zeros((W+1, n+1))
        for i in range(1, W+1):
            for j in range(1, n+1):
                value[i][j] = value[i][j-1]
                if iw[j-1]<=i:
                    temp = value[i-iw[j-1]][j-1] + iw[j-1]
                    if temp > value[i][j]:
                        value[i][j] = temp
                if value[i][j] == W:
                    count += 1

        if count < 3:
            print('0')
        else:
            print('1')

if __name__ == '__main__':
    n = int(input())
    iw = [int(i) for i in input().split()]
    solution(n,iw)
```

## 3. Maximum Value of an Arithmetic Expression

```python
import math

def calculate(a, b, c):

    if c == '+':
        return a + b
    elif c == '-':
        return a - b
    else:
        return a * b

def Min_Max(M, m, i, j, operators):
    min_value = math.inf
    max_value = -math.inf

    for k in range(i, j):
        a = calculate(M[i][k], M[k+1][j], operators[k])
        b = calculate(M[i][k], m[k+1][j], operators[k])
        c = calculate(m[i][k], M[k+1][j], operators[k])
        d = calculate(m[i][k], m[k+1][j], operators[k])
        min_value = min(min_value, a, b, c, d)
        max_value = max(max_value, a, b, c, d)
        
    return [min_value, max_value]

def solution(exp, operators, operands):

    n = len(operands)
    m = [[None for x in range(n)] for x in range(n)]
    M = [[None for x in range(n)] for x in range(n)]

    for i in range(n):
        m[i][i] = operands[i]
        M[i][i] = operands[i]

    for s in range(1, n):
        for i in range(0, n-s):
            j = i + s
            res = Min_Max(M, m, i, j, operators)
            m[i][j] = res[0]
            M[i][j] = res[1]

    print(M[0][n-1])

if __name__ == '__main__':
    exp = input()
    operators, operands = [], []
    for i in exp:
        if i in ['+', '-', '*']:
            operators.append(i)
        else:
            operands.append(int(i))
    solution(exp, operators, operands)
```
